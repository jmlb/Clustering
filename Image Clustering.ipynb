\sum (x - \mu)^2

```python
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import pandas as pd
import math as math
```

```python
######## INPUT SECTION ###############
#img = Image.open("TG Particle_NEW_01A.jpg")
img = Image.open("download.jpg")
Nbr_of_centroid = 6
do_crop=False
forced_centroids = np.array([[255/255.0,255/255.0,255/255.0],[0, 0, 0]])
force_centroids = np.array([])
take_slab = False
########################################

if len(forced_centroids) > Nbr_of_centroid:
    print('Warning: Nbr_of_centroids (', Nbr_of_centroid, ') \
    < Nbr_of_forced_centroids(',len(forced_centroids),')')
    print('\n Nbr of centroids has been set to Nbr of Forced_Centroids.')


if do_crop:
    img = img.crop((15, 37, 362, 384)) #crop the image to extract disk only
width, height = img.size  #size of cropped image
imgMatrix = np.array(img)

print ('******* Original Image **********')
plt.figure(1)
img_png = Image.fromarray(imgMatrix, 'RGB')
plt.imshow(img_png)
```

## Functions required for optimization:

assign_cluster()

initialize_centroid()

cost_function() [calculate cost function at all iterations]

show_centroidsColor() [show color of final centroids]

Mask Generator: only 

```python

def assign_cluster(df_X, centroid_array):
    #pd_X is a dataframe with the training examples with cols=R, G, B, class
    #array_centroid = list of centroids is a matrix size=(K_centroid, n_features)
    Nbr_of_centroids = centroid_array.shape[0]
    Nbr_of_samples = df_X.shape[0]
    #for every sample, the distance to each centroid is calculated
    distance_to_centroid = np.zeros((Nbr_of_samples, Nbr_of_centroids)) 
    df_X = df_X[['R','G','B']] #select RGB columns only
    X_array = df_X.as_matrix()
    for jj in range(0, Nbr_of_centroids): #for each centroid, calculate the distance squared sample to centroid
        this_centroid = centroid_array[jj]
        diffSqr = np.sum((X_array - this_centroid)*(X_array - this_centroid), axis=1)
        distance_to_centroid[:,jj] = np.transpose(diffSqr) #get a vector
    return [np.argmin(distance_to_centroid, axis=1), distance_to_centroid]

def initialize_centroid(df_X, centroid_array, f_centroid):    
    # Get mean of all samples from same cluster assignement
    Nbr_of_centroids = centroid_array.shape[0]
    if len(f_centroid) == 0:
        for jj in range(0, Nbr_of_centroids):
            bucket = df_X[df_X['class']==jj][['R','G','B']]
            mean_bucket = bucket.sum(axis=0 )/(bucket.shape[0]*1.0)
            centroid_array[jj,:] = mean_bucket.as_matrix()
    elif (len(f_centroid) < Nbr_of_centroids):
        for jj in range(len(f_centroid), Nbr_of_centroids):
            bucket = df_X[df_X['class']==jj][['R','G','B']]
            mean_bucket = bucket.sum(axis=0 )/(bucket.shape[0]*1.0)
            centroid_array[jj,:] = mean_bucket.as_matrix()
        
    return centroid_array

def cost_function(dist):
    #dist is a numpy array [dist_centroid0, dist_centroid1, ..., centroid5]
    min_dist = np.min(dist, axis=1)
    sum_of_distance = np.sum(min_dist)
    Nbr_of_samples = dist.shape[0]
    return sum_of_distance/Nbr_of_samples

#Optional functions
def show_CentroidsColor(centroids, iter):
    box=np.ones((100,1,3))
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for index, centroid in enumerate(centroids):
        pattern = np.ones((100,100,3))*centroid*255
        box = np.hstack((box,pattern))
        ax.text(index*100, -10, (centroid*255).astype(int), fontsize=8)
        title = 'Vizualization of Centroids at iteration:', iter
        ax.text(100, -50, title, fontsize=15)
        plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')
        plt.tick_params(axis='y', which='both', left='off', right='off', labelleft='off')   
    plt.imshow(box.astype('uint8'))
```

```python
#need to normalize the array to the max value (255) so the input is between 0-1 (255.0 to get float numbers)
imgMatrix = imgMatrix/255.0
height, width, depth =  np.shape(imgMatrix) #dimensions of the matrix (height, )
unwind_imgMatrix = np.reshape(imgMatrix,(width*height,3)) 
#reshape the matrix into m*3 matrix where 3 si for 
#the values RGB  [R G B]

colName = ['R', 'G', 'B']
pd_Data = pd.DataFrame(unwind_imgMatrix, columns=colName)
pd_Data['class'] = np.ones((width*height,1))*-1

print('# of training examples (initial)', pd_Data.shape[0])
pd_Data_clean = pd_Data.drop_duplicates()

data_clean_size = pd_Data_clean.shape[0]

#Plot few examples
fig = plt.figure(5)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(unwind_imgMatrix[0:1000,0],unwind_imgMatrix[0:1000,1],unwind_imgMatrix[0:1000,2], c='r', marker='o')
ax.set_title('3D Vizualization of a few RGB points')
ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')

#set Initial value centroids
#randomly pick 6 indexes
centroids_idx = np.zeros((Nbr_of_centroid, 1), dtype=int)
id_seed = np.arange(data_clean_size)
np.random.shuffle(id_seed)
centroids_idx = id_seed[0:Nbr_of_centroid]
centroid_set = pd_Data_clean.iloc[centroids_idx][['R', 'G','B']] #(6*3) matrix
centroid_set = centroid_set.as_matrix()

if len(forced_centroids) == 0 or len(forced_centroids) > Nbr_of_centroid:
    print('Can`t do that: the Nbr of forced Centroids must not be larger \
    than the total Nbr of class/centroid')
else:
    for id, forced_item in enumerate(forced_centroids):
        centroid_set[id] = forced_item
max_iter = 10
cost = np.zeros((max_iter,2))

for iter in range(max_iter):
    assign_cluster_res =  assign_cluster(pd_Data_clean, centroid_set)
    pd_Data_clean['class'] = assign_cluster_res[0]
    cost[iter] = [iter, cost_function(assign_cluster_res[1])]
    centroid_set = initialize_centroid(pd_Data_clean, centroid_set, forced_centroids)
    show_CentroidsColor(centroid_set, iter)
```

```python
#Plot cost_function versus iteration
fig = plt.figure(2)
ax = fig.add_subplot(111)
ax.scatter(cost[:,0],cost[:,1], c='r', marker='o', s=50 )
ax.set_title('Cost Function versus Iteration')
ax.set_xlabel('Number of Iterations ')
ax.set_ylabel('Cost Function (arb. unit)')
```

```python
#Go from pd_data_clean to pd_data with all classes
pd_Data = pd_Data.join(pd_Data_clean['class'], rsuffix='_new')
pd_Data_sort = pd_Data.sort_values(by=['R','G', 'B'])
pd_Data_sort['class_new'] = \
pd_Data_sort['class_new'].replace('NaN', np.nan).fillna(method='ffill').astype(int)
pd_Data_sort.sort_index(inplace=True)
newImg = pd_Data_sort.as_matrix(['R', 'G', 'B'])

#change RGB of examples to RGB of assigned centroid
for kk in range(centroid_set.shape[0]):
    elt = pd_Data_sort[pd_Data_sort['class_new'] == kk].index.values
    pd_Data_sort.set_value(elt, 'new_R', (centroid_set[kk][0]*255).astype(int))
    pd_Data_sort.set_value(elt, 'new_G', (centroid_set[kk][1]*255).astype(int))
    pd_Data_sort.set_value(elt, 'new_B', (centroid_set[kk][2]*255).astype(int))
newImg = pd_Data_sort.as_matrix(['new_R', 'new_G', 'new_B'])
reshape_img = np.reshape(newImg, (height, width, 3))
newImg_png = Image.fromarray(reshape_img.astype('uint8'), 'RGB')

if take_slab == True:
    #get radial defect coverage
    x0, y0 = [math.floor(height/2), math.floor(width/2)]
    radiusID_mm = 22
    radiusOD_mm = 27
    disk_diameter_mm = 65
    radiusID_px = math.floor(height/disk_diameter_mm*radiusID_mm)
    radiusOD_px = math.floor(height/disk_diameter_mm*radiusOD_mm)

    xx, yy = np.mgrid[0:height,0:width]
    radius = ((xx-x0)**2 + (yy-y0)**2)**0.5
    # circular mask
    circmaskID = (radius >=radiusID_px)
    circmaskOD = (radius <=radiusOD_px)
    circmask = circmaskID * circmaskOD
    reshape_img[~circmask] = 0
    newImg_png_slice = Image.fromarray(reshape_img.astype('uint8'))
    print('******* IMages **********')
    plt.imshow(newImg_png_slice)

## count pixels of each class in 
summary_pxCount = np.zeros((Nbr_of_centroid, 4), dtype=int)
np_snapshot = np.reshape(reshape_img, (height*width,3))
for kk in range(centroid_set.shape[0]):
    centroid = (centroid_set[kk]*255).astype(int)
    pixel_count = ((np_snapshot == centroid).all(axis=1)).sum()
    data_cent = np.append(centroid, pixel_count)
    summary_pxCount[kk] = data_cent.astype(int)
print('******* IMages **********')
plt.imshow(newImg_png)
print('************* Summary Table *********\n')
print(' R|G|B|Nbr of Pixels')
print(summary_pxCount)
```
